```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```
print(df)
```

# Simple Bar charts

```
fig = plt.figure(figsize =(20, 10)) 
ax = sns.barplot(x="Year", y="Freq", hue="Product", data=df)
```

```
g = sns.catplot(x="Year", y="Freq", col="Product", data=df, kind="bar", height=10, aspect=.7);
```

```
g = sns.catplot(x="Year", y="Freq", col="Product", col_wrap=2,
                data=df, kind="bar", height=5, aspect=.8)
```

```
g = sns.catplot(x="Year", y="Freq", col="Product", col_wrap=2, palette = 'magma',
                data=df, kind="bar", height=5, aspect=.8)
```

# Advanced Bar Charts Using Plotly

```
import plotly.express as px
fig = px.bar(df, x="Year", y="Freq", color="Product", title="Bar chart with all Days Values")
fig.show()
```

```
import plotly.express as px
fig = px.bar(df,  x="Year", y="Freq", color="Product", title="Bar chart with all Days Values",
             hover_data=['Freq', "Product"],  height=400)
fig.show()
```
