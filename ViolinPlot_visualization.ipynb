```
import pandas as pd
df1 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Datasets/Wajdi visualisations/Boxplot/Android_FT_A11yGap.csv')
df2 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Datasets/Wajdi visualisations/Boxplot/Android_FT_NON_A11yGap.csv')
df3 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Datasets/Wajdi visualisations/Boxplot/iOS_FT_A11yGap.csv')
df4 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Datasets/Wajdi visualisations/Boxplot/iOS_FT_NON_A11yGap.csv')
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

# Violinplot Visualisation

```
fig = plt.figure(figsize=(20, 8))
gs = fig.add_gridspec(1, 4)

ax = fig.add_subplot(gs[0, 0])

sns.violinplot(data=df1.Days,palette= "Set2" )
ax.set_xlabel("Android A11y")

ax = fig.add_subplot(gs[0, 1])
sns.violinplot(data=df2.Days, palette= "Paired")
ax.set_xlabel("Android Non−A11y")

ax = fig.add_subplot(gs[0, 2])
sns.violinplot(data=df3.Days, palette= "flare")
ax.set_xlabel("iOS A11y")

ax = fig.add_subplot(gs[0, 3])
sns.violinplot(data=df4.Days, palette= "crest")
ax.set_xlabel("iOS Non−A11y")



fig.tight_layout()
plt.title("Life Expectancy By Country")
plt.show()
```

```
fig = plt.figure(figsize=(10, 5))
gs = fig.add_gridspec(1, 4)

ax = fig.add_subplot(gs[0, 0])

sns.violinplot(data=df1.Days,palette= "Set2" )
sns.swarmplot(data=df1.Days, color="k", alpha=0.3)
ax.set_xlabel("Android A11y")

ax = fig.add_subplot(gs[0, 1])
sns.violinplot(data=df2.Days, palette= "Paired")
sns.swarmplot(data=df2.Days, color="k", alpha=0.3)
ax.set_xlabel("Android Non−A11y")

ax = fig.add_subplot(gs[0, 2])
sns.violinplot(data=df3.Days, palette= "flare")
sns.swarmplot(data=df3.Days, color="k", alpha=0.3)
ax.set_xlabel("iOS A11y")

ax = fig.add_subplot(gs[0, 3])
sns.violinplot(data=df4.Days, palette= "crest")
sns.swarmplot(data=df4.Days, color="k", alpha=0.3)
ax.set_xlabel("iOS Non−A11y")
plt.title("Violinplot with all Days Values")

fig.tight_layout()
plt.show()
```

```
fig = plt.figure(figsize=(20, 8))
gs = fig.add_gridspec(1, 4)

ax = fig.add_subplot(gs[0, 0])
sns.violinplot(data=df1.Days,palette= "Set2" , cut=0)
plt.boxplot(df1.Days, )
ax.set_xlabel("Android A11y")

ax = fig.add_subplot(gs[0, 1])
sns.violinplot(data=df2.Days, palette= "Paired", cut=0)
plt.boxplot(df2.Days, )
ax.set_xlabel("Android Non−A11y")

ax = fig.add_subplot(gs[0, 2])
sns.violinplot(data=df3.Days, palette= "flare", cut=0)
plt.boxplot(df3.Days, )
ax.set_xlabel("iOS A11y")

ax = fig.add_subplot(gs[0, 3])
sns.violinplot(data=df4.Days, palette= "crest", cut=0)
plt.boxplot(df4.Days, )
ax.set_xlabel("iOS Non−A11y")

fig.tight_layout()
plt.show()
```

```
fig = plt.figure(figsize=(20, 8))
gs = fig.add_gridspec(1, 4)

ax = fig.add_subplot(gs[0, 0])
sns.violinplot(y = df1.Days,palette= "Set2" , cut=0, orient= "v")
ax.set_xlabel("Android A11y")

ax = fig.add_subplot(gs[0, 1])
sns.violinplot(y = df2.Days, palette= "Paired", cut=0, orient= "v")
ax.set_xlabel("Android Non−A11y")

ax = fig.add_subplot(gs[0, 2])
sns.violinplot(y = df3.Days, palette= "flare", cut=0, orient= "v")
ax.set_xlabel("iOS A11y")

ax = fig.add_subplot(gs[0, 3])
sns.violinplot(y = df4.Days, palette= "crest", cut=0, orient= "v")
ax.set_xlabel("iOS Non−A11y")

fig.tight_layout()
plt.show()
```

```
fig = plt.figure(figsize=(20, 8))
gs = fig.add_gridspec(1, 4)

ax = fig.add_subplot(gs[0, 0])
ax = sns.violinplot(y=df1.Days,palette= "Set2" ,inner=None, color=".8", orient= "v", cut=0)
ax = sns.stripplot(y= df1.Days)
ax.set_xlabel("Android A11y")

ax = fig.add_subplot(gs[0, 1])
ax = sns.violinplot(y = df2.Days, palette= "Paired",inner=None, color=".8", orient= "v", cut=0)
ax = sns.stripplot(y = df2.Days)
ax.set_xlabel("Android Non−A11y")

ax = fig.add_subplot(gs[0, 2])
ax = sns.violinplot(y = df3.Days, palette= "flare",inner=None, color=".8", orient= "v", cut=0)
ax = sns.stripplot(y = df3.Days)
ax.set_xlabel("iOS A11y")

ax = fig.add_subplot(gs[0, 3])
ax = sns.violinplot(y = df4.Days, palette= "crest",inner=None, color=".8", orient= "v", cut=0)
ax = sns.stripplot(y = df4.Days)
ax.set_xlabel("iOS Non−A11y")

fig.tight_layout()
plt.show()
```
