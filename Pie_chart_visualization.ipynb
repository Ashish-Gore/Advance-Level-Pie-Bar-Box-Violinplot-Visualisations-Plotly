```
import plotly.graph_objects as go
from plotly.subplots import make_subplots

labels = ['2014', '2015', '2016', '2017','2018', '2019','2020']

# Create subplots: use 'domain' type for Pie subplot
fig = make_subplots(rows=1, cols=4, specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}, {'type':'domain'}]])
fig.add_trace(go.Pie(labels=labels, values=[1030,1804,2260,6701,6250,5970,4943], name="Non-A11y Android"),
              1, 1)
fig.add_trace(go.Pie(labels=labels, values=[296,374,441,1986,1784,1459,1375], name="Non-A11y iOS"),
              1, 2)
fig.add_trace(go.Pie(labels=labels, values=[203,200,133,109,102,139,133], name="Non-A11y Android"),
              1, 3)
fig.add_trace(go.Pie(labels=labels, values=[156,156,66,72,55,67,31], name="Non-A11y iOS"),
              1, 4)

# Use `hole` to create a donut-like pie chart
fig.update_traces(hole=.4, hoverinfo="label+percent+name")

fig.update_layout(
    title_text="Pie chart with all Days Values",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Non-A11y Android', x=0.06, y=0.5, font_size=10, showarrow=False),
                 dict(text='Non-A11y iOS', x=0.37, y=0.5, font_size=10, showarrow=False),
                 dict(text='A11y Android',  x=0.630, y=0.5, font_size=10, showarrow=False),
                 dict(text='A11y iOS', x=0.92, y=0.5, font_size=10, showarrow=False)])
fig.show()
```

```
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import numpy as np

# make figure and assign axis objects
fig = plt.figure(figsize=(18, 10.125))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
fig.subplots_adjust(wspace=0)

# large pie chart parameters
ratios = [.201, .756, .0266,.0157]
labels = ['Non-A11y iOS', 'Non-A11y Android',  'A11y Android', 'A11y iOS']
explode = [0.1, 0, 0,0]
# rotate so that first wedge is split by the x-axis
angle = -180 * ratios[0]
ax1.pie(ratios, autopct='%1.1f%%', startangle=angle,
        labels=labels, explode=explode)

# small pie chart parameters
ratios = [296,374,441,1986,1784,1459,1375]
labels = ['2014', '2015', '2016', '2017','2018', '2019','2020']
width = .2

ax2.pie(ratios, autopct='%1.1f%%', startangle=angle,
        labels=labels, radius=0.5, textprops={'size': 'smaller'})

ax1.set_title('Product')
ax2.set_title('Year wise Days')

# use ConnectionPatch to draw lines between the two plots
# get the wedge data
theta1, theta2 = ax1.patches[0].theta1, ax1.patches[0].theta2
center, r = ax1.patches[0].center, ax1.patches[0].r

# draw top connecting line
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(- width / 2, .5), xyB=(x, y),
                      coordsA="data", coordsB="data", axesA=ax2, axesB=ax1)
con.set_color([0, 0, 0])
con.set_linewidth(2)
ax2.add_artist(con)

# draw bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(- width / 2, -.5), xyB=(x, y), coordsA="data",
                      coordsB="data", axesA=ax2, axesB=ax1)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(2)

plt.show()
```
