# Boxplot Visualisation

## With Different Dimensions

```
data = [df1.Days, df2.Days, df3.Days, df4.Days]
fig = plt.figure(figsize =(20, 10)) 
ax = fig.add_subplot(111) 
  
# Creating axes instance 
bp = ax.boxplot(data, patch_artist = True, vert = 0) 

# y-axis labels 
ax.set_yticklabels(['Android A11y', 'Android Non−A11y', 'iOS A11y', 'iOS Non−A11y']) 
  
# Adding title  
plt.title("Boxplot with all Days Values") 
  
# Removing top axes and right axes ticks 
ax.get_xaxis().tick_bottom() 
ax.get_yaxis().tick_left() 
      
# show plot 
plt.show(bp) 
```

## With Different Palette color

```
sns.set_style("whitegrid")
sns.color_palette("crest", as_cmap=True)


data = [df1.Days, df2.Days, df3.Days, df4.Days]
fig = plt.figure(figsize =(10, 10)) 
ax = fig.add_subplot(111) 
  
# Creating axes instance 
bp = ax.boxplot(data, patch_artist = True, vert = 1) 

colors = ['#0000FF', '#00FF00',  
          '#FFFF00', '#FF00FF'] 
  
for patch, color in zip(bp['boxes'], colors): 
    patch.set_facecolor(color) 

# x-axis labels 
ax.set_xticklabels(['Android A11y', 'Android Non−A11y', 'iOS A11y', 'iOS Non−A11y']) 
  
# Adding title  
plt.title("Boxplot with all Days Values") 
  
# Removing top axes and right axes ticks 
ax.get_xaxis().tick_bottom() 
ax.get_yaxis().tick_left() 
      

# show plot 
plt.show(bp) 
```
